#**Exploratory Data Analysis(EDA) of NASA Dataset**

## **What is Exploratory Data Analysis?**

**Exploratory Data Analysis refers to the critical process of performing initial investigations on data so as to discover patterns,
to spot anomalies,
to test hypothesis and to check assumptions with the help of summary statistics and graphical representations.**

**It is a good practice to understand the data first and try to gather as many insights from it. EDA is all about making sense of data in hand,before getting them dirty with it.**

### **Induidual Variable Analysis**

```python
numer_cols = ['absolute_magnitude_h', 'kilometers_estimated_diameter_min', 'kilometers_estimated_diameter_max', 'mean_esitmated_diameter(Km)', 'perihelion_distance', 'aphelion_distance', 'first_observation_year', 'last_observation_year']
cat_cols = ['Name', 'is_potentially_hazardous_asteroid', 'is_sentry_object', 'orbit_class_type']

for col in numer_cols:
  sb.histplot(data=data, x=col, multiple='dodge', stat='probability', kde=True)
  plt.show()

```

```python
import matplotlib
plt.style.use("ggplot")
matplotlib.rc('xtick', labelsize=20)
matplotlib.rc('ytick', labelsize=20)
sb.countplot(data=data, x='first_observation_year', hue='is_potentially_hazardous_asteroid', palette=['#0EFF14','#FF2900'])
plt.xlabel("First Observation Year", fontsize=30, fontweight='bold')
plt.xticks(np.arange(1,34,5))
plt.ylabel("Number of Asteroids Discovered", fontsize=30, fontweight='bold')
plt.legend(['Non-Hazardous', 'Hazardous'],fontsize=20)
plt.show()
```

```python
plt.style.use('ggplot')
for col in cat_cols:
  if col != 'Name':
    print(data[col].value_counts())
    data[col].value_counts().plot.pie(autopct='%1.2f%%', shadow=True, startangle=30,textprops={'fontsize': 25, 'fontweight':'bold'}, colors=sb.color_palette('Set2'), labeldistance=0.75,
                                                 pctdistance=0.55)
    plt.show()

```

```python
for col in numer_cols:
  sb.boxplot(data=data, y=col, showfliers=False)
  plt.show()
```

### **Paired Variable Analysis**

```python
for col in numer_cols:
  sb.histplot(data=data, x=col, hue='is_potentially_hazardous_asteroid',multiple='dodge', kde=True)
  plt.show()
```

```python
lst = numer_cols
lst.append('is_potentially_hazardous_asteroid')

sb.pairplot(data=data[lst], hue = 'is_potentially_hazardous_asteroid')
plt.show()
```

```python
import matplotlib
matplotlib.rc('xtick', labelsize=15)
matplotlib.rc('ytick', labelsize=15)

fig = plt.figure(figsize=(20,15))
sb.catplot(data=data, x="is_potentially_hazardous_asteroid", kind="count", hue='is_sentry_object', palette=['#17009B','#BA0000'])


plt.xlabel("Hazardous Asteroid", fontsize=20, fontweight='bold')
plt.ylabel("Count frequency", fontsize=20, fontweight='bold')

plt.show()

df = data[data['is_sentry_object'] == True]
df['is_potentially_hazardous_asteroid'].value_counts()
```

*  **Mostly Sentry objects are not potentially hazardous.**




```python
data['first_observation_year'].sort_values(ascending=True).value_counts().plot(kind='bar')
plt.show()
```

```python
data['last_observation_year'].value_counts().plot(kind='bar')
plt.show()
```

*   **Most of the asteroids were discovered in recent years.**




```python
for col in cat_cols:
  if col != 'Name':
    sb.boxplot(data=data, x=col, y='perihelion_distance')
    plt.show()
```

```python
for col in cat_cols:
  if col != 'Name':
    sb.violinplot(data=data, x=col, y='perihelion_distance')
    plt.show()
```

### **Correlation Analysis**

```python
mask=np.triu(np.ones_like(data.corr("spearman"),dtype=bool))
sb.heatmap(data.corr(),annot=True,mask=mask,cmap= "vlag")

plt.show()
```
